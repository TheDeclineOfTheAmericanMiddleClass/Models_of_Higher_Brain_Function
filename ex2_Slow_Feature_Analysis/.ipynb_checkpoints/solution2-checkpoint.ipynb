```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
```

**Exercise 1: Slow Feature Analysis on two independent signals**

```python
# 1.1
def psd(k, epsilon):
    return np.exp(-np.abs(k)**2/(2*epsilon))

def random_signal(T, epsilon):
    random_signal = np.random.rand(T)
    fft = np.fft.fft(random_signal)
    k = np.fft.fftfreq(T)
    
    psd_ = psd(k, epsilon)
    psd_normalized = np.sqrt(psd_)
    psd_normalized /= np.sum(psd_normalized)
    
    fft_filtered = fft * psd_normalized
    filtered_random_signal = np.real(np.fft.ifft(fft_filtered))
    return filtered_random_signal

fig, axs = plt.subplots(1, 4, figsize=(16, 3))
for i, epsilon in enumerate([0.1, 0.01, 0.001, 0.0001]):
    axs[i].plot(random_signal(200, epsilon))
    axs[i].set_title(f"Signal ($\epsilon$ = {epsilon})")
    axs[i].set_xlabel("Time")
    axs[i].set_ylabel("Value")
fig.tight_layout()
```

```python
# 1.2
T = 1000
tau_1 = 30
tau_2 = 100
epsilon_1 = 1/tau_1**2
epsilon_2 = 1/tau_2**2

# Normalizing to zero mean and unit variance
signal_1 = random_signal(T, epsilon_1) - np.mean(random_signal(T, epsilon_1))
signal_2 = random_signal(T, epsilon_2) - np.mean(random_signal(T, epsilon_2))

signal_1 /= np.std(signal_1)
signal_2 /= np.std(signal_2)

fig, ax = plt.subplots()
ax.plot(signal_1, label=f"Signal 1 ($\\tau$ = {tau_1})")
ax.plot(signal_2, label=f"Signal 2 ($\\tau$ = {tau_2})")
ax.set_title("Signals")
ax.set_xlabel("Time")
ax.set_ylabel("Value")
ax.legend()
fig.tight_layout()
```

With lower $\tau$, and therefore bigger the $\epsilon$, high-frequencies are filtered out more and more of the total power is used for low frequencies. As a consequence, signal 2 is  smoother than signal 1.

```python
# 1.3
X = np.stack([signal_1, signal_2])
C = np.cov(X)
X_dot = np.diff(X, axis=1)
C_dot = X_dot@X_dot.T/(T-1)
eigenvalues, eigenvectors = linalg.eig(C_dot, C)

print(f"Normalized: {np.linalg.norm(eigenvectors, axis=0)}")
```

```python
# 1.4
for i, ev in enumerate(eigenvectors.T):
    print(f"Eigenvector {i+1}: {ev} with eigenvalue {eigenvalues[i]}")
```

Given how the data was generated, we would expect that there is (almost) no mixing, since both signals have been independently generated from independently filtered white noise and thus there is no way to compensate for the temporal variance in one of the signals by combining it with the other.

Empirically, this assumption is validated. The weights (eigenvectors) for the optimal function for SFA predominantly select either signal 1 or signal 2 respectively. Yet, the values for signal 1 are negative.

```python
mixing = eigenvectors@X

fig, ax = plt.subplots()
ax.plot(mixing.T)
ax.set_title("Optimal functions for SFA")
ax.set_xlabel("Time")
ax.set_ylabel("Value")
fig.tight_layout()
```

```python
def generate_sfa_weights(T, tau_1, tau_2):
    signal_1 = random_signal(T, 1/tau_1**2)
    signal_2 = random_signal(T, 1/tau_2**2)

    # # Normalizing to zero mean and unit variance
    # signal_1 = random_signal(T, epsilon_1) - np.mean(random_signal(T, epsilon_1))
    # signal_2 = random_signal(T, epsilon_2) - np.mean(random_signal(T, epsilon_2))

    # signal_1 /= np.std(signal_1)
    # signal_2 /= np.std(signal_2)
    
    X = np.stack([signal_1, signal_2])
    C = np.cov(X)
    X_dot = np.diff(X, axis=1)
    C_dot = X_dot@X_dot.T/(T-1)
    eigenvalues, eigenvectors = linalg.eig(C_dot, C)
    return eigenvectors, eigenvalues

samples_n = 20
sample_sizes = np.logspace(1, 4, samples_n).astype(np.int32)
all_weights = np.zeros((samples_n, 4))
for i, sample_size in enumerate(sample_sizes):
    weights, _ = generate_sfa_weights(sample_size, tau_1, tau_2)
    all_weights[i, :] = weights.T.flatten()

fig, ax = plt.subplots()
ax.semilogx(sample_sizes, all_weights[:, 0], c="red", linestyle="-", label="$W_{00}$")
ax.semilogx(sample_sizes, all_weights[:, 1], c="red", linestyle="--", label="$W_{01}$")
ax.semilogx(sample_sizes, all_weights[:, 2], c="green", linestyle="-", label="$W_{10}$")
ax.semilogx(sample_sizes, all_weights[:, 3], c="green", linestyle="--", label="$W_{11}$")

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax.set_title("SFA-mixture for different signal lenghts")
ax.set_xlabel("Signal Length")
ax.set_ylabel("Weight Value")
fig.tight_layout()
```

The longer the signal length, the more strictly the signals are independently selected without mixture. For short signal lengths, the mixture varies for different samples. This is due to the fact that for short signals, temporal correlations between the signals can occur by chance, which then can be used to reduce temporal variations within the mixture.

```python
# 1.6
T = 10000
all_eigenvalues_1 = np.zeros((100, 100))
all_eigenvalues_2 = np.zeros((100, 100))
for i, tau_1 in enumerate(np.linspace(1, 30, 100)):
    for n, tau_2 in enumerate(np.linspace(1, 30, 100)):
        _, eigenvalues = generate_sfa_weights(T, tau_1, tau_2)
        all_eigenvalues_1[i, n] = np.real(eigenvalues[0])
        all_eigenvalues_2[i, n] = np.real(eigenvalues[1])
```

```python

from mpl_toolkits.axes_grid1 import make_axes_locatable
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
fig.subplots_adjust(left=0.02, bottom=0.06, right=0.95, top=0.94, wspace=0.05)

im0 = axs[0].matshow(all_eigenvalues_1*all_eigenvalues_2)
axs[0].xaxis.tick_bottom()
axs[0].set_title("First Eigenvalue")
axs[0].set_xlabel(f"Tau 2")
axs[0].set_ylabel(f"Tau 1")

# Clean, cool colorbar
divider = make_axes_locatable(axs[0])
cax0 = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(im0,ax=axs[0],cax=cax0)

im1 = axs[1].matshow(all_eigenvalues_2)
axs[1].xaxis.tick_bottom()
axs[1].set_title("Second Eigenvalue")
axs[1].set_xlabel(f"Tau 2")
axs[1].set_ylabel(f"Tau 1")

divider = make_axes_locatable(axs[1])
cax1 = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(im1, ax=axs[1], cax=cax1)

fig.tight_layout()
```

Only if the signals are different ($\tau_1$ is unqual to $\tau_2$), we can benefit from mixing the signals with unqual contributions (the eigenvalues are different). Whereas the signals are mixed with equal contribution, if the $\tau$s are identical.

**Exercise 2: SFA on a high-dimensional correlated signal**

```python
# 2.1
L = 10000
epsilon = 0.04
signal = random_signal(L, epsilon)

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(signal)
ax.set_title("Signal")
ax.set_xlabel("Time")
ax.set_ylabel("Value")
fig.tight_layout()
```

```python
# 2.2
window_width = 100
X = np.zeros((window_width, L-window_width))
for step in range(L-window_width):
    X[:, step] = signal[step:(step+window_width)]
X = (X.T-np.mean(X, axis=1)).T
```

```python
# 2.3
C = np.cov(X)
X_dot = np.diff(X, axis=1)
C_dot = X_dot@X_dot.T/(T-1)
eigenvalues, eigenvectors = linalg.eig(C_dot, C)
print(eigenvalues.shape, eigenvectors.shape)
```

```python
# 2.4
smallest_n = 5
idx = np.argsort(np.abs(eigenvalues))
W = eigenvectors[:, idx[:smallest_n]]

fig, ax = plt.subplots()
ax.plot(W)
ax.set_title("Eigenvectors with slowest SFA component")
ax.set_xlabel("Filter position")
ax.set_ylabel("Value")
fig.tight_layout()
```

#### 2.5
The eigenvectors represent sinusoids of increasing frequency. If we interpret them as temporal receptive fields, we can see that RFs of increasing eigenvalues respond to inputs of higher and higher temporal frequencies. Thus, the receptive field of the first eigenvector (blue) is sensitive to the slowest changes in the signal, the second eigenvector (orange) to slightly faster changes, and so on -- the intention of SFA. 

The eigenvectors tend further away from 0 at the boundaries of the receptive field, as their respective eigenvalues increase. If we consider edges in images are places of high spatial frequency, it makes sense that eigenvectors with higher eigenvalues, which code for faster spatial features like edges, are have higher weights at the receptive field boundaries.

If we interpret the eigenvectors as spatial receptive fields, they are responsive to distinct spatial frequencies, similar to the receptive fields of V1 neurons, which are often modeled by a gabor function, which, in addition to the sinusoid, includes a gaussian to limit the spatial extent of the receptive field.

```python
# 2.6
Y = W.T@X
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(signal, label="Signal")
for i, offset in enumerate(np.linspace(0.0005, 0.002, smallest_n)):
    ax.plot(Y.T[:, i] + offset, label=f"FSA Component {i+1}")

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax.set_title("FSA Result")
ax.set_xlabel("Time")
ax.set_ylabel("(Shifted) Values")

fig.tight_layout()
```

Yes, the extracted components change a lot slower than the original signal.
